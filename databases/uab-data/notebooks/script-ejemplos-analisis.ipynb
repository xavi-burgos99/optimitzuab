### 3. Identificación y Manejo de Valores Nulos

- Identifica los valores nulos en los datasets grupos y ubicaciones.
- Muestra cómo imputar valores faltantes utilizando otros campos o estadísticas (e.g., media).
- Los participantes deben decidir la mejor estrategia de imputación según el contexto.

```python
# Verificar valores nulos en 'grupos'
print("Valores nulos en 'grupos':")
print(grupos.isnull().sum())

# Imputar valores nulos en IND_ALUMNOS_GRUPO_REAL con IND_ALUMNOS_GRUPO_PREV
grupos['IND_ALUMNOS_GRUPO_REAL'].fillna(grupos['IND_ALUMNOS_GRUPO_PREV'], inplace=True)

# Verificar valores nulos en 'ubicaciones'
print("\nValores nulos en 'ubicaciones':")
print(ubicaciones.isnull().sum())

# Imputar valores nulos en 'IND_METRES' con la media
ubicaciones['IND_METRES'].fillna(ubicaciones['IND_METRES'].mean(), inplace=True)
```

### 4. Estandarización de Tipos de Datos

- Asegura que los campos identificadores están en el tipo de datos correcto para facilitar las uniones.
- Convierte campos de fecha al tipo datetime para permitir operaciones basadas en tiempo.
- Los participantes deben manejar posibles errores o valores no convertibles.

```python
# Asegurar que los identificadores clave son de tipo string
grupos['ID_GRUPO'] = grupos['ID_GRUPO'].astype(str)
ubicaciones['ID_UBICACIO'] = ubicaciones['ID_UBICACIO'].astype(str)
calendario_grupos['ID_GRUPO'] = calendario_grupos['ID_GRUPO'].astype(str)
calendario_grupos['ID_UBICACION'] = calendario_grupos['ID_UBICACION'].astype(str)

# Convertir fechas a formato datetime
calendario_grupos['ID_FECHA_GRUPO'] = pd.to_datetime(calendario_grupos['ID_FECHA_GRUPO'], errors='coerce')

# Verificar conversiones
print("\nTipos de datos en calendario_grupos:")
print(calendario_grupos.dtypes)
```

### 5. Unión de Datasets

- Muestra cómo unir los datasets clave para crear un conjunto de datos consolidado que permita el análisis.
- Utiliza uniones left para conservar todos los registros de calendario_grupos y agregar información de otros datasets.
- Los participantes deben verificar que las uniones sean correctas y manejar registros sin correspondencia.

```python
# Unir 'calendario_grupos' con 'grupos' para obtener información sobre el número de estudiantes
calendario_completo = calendario_grupos.merge(
    grupos[['ID_GRUPO', 'IND_ALUMNOS_GRUPO_REAL']], 
    on='ID_GRUPO', 
    how='left'
)

# Unir con 'ubicaciones' para obtener información sobre las aulas
calendario_completo = calendario_completo.merge(
    ubicaciones[['ID_UBICACIO', 'IND_METRES']], 
    left_on='ID_UBICACION',
    right_on='ID_UBICACIO',
    how='left'
)

# Verificar el resultado
print("Calendario completo:")
display(calendario_completo.head())
```

### 6. Validación de Restricciones Básicas

- Identifica casos donde el número de estudiantes excede la capacidad del aula.
- Calcula el porcentaje de ocupación de cada clase.
- Detecta posibles solapamientos de horarios en una misma aula.
- Los participantes pueden profundizar en estas validaciones y corregir inconsistencias.

```python
calendario_completo['CAPACIDAD_ESTIMADA'] = (calendario_completo['IND_METRES_y'] / 2).round()
calendario_completo['EXCEDE_CAPACIDAD'] = calendario_completo['IND_ALUMNOS_GRUPO_REAL'] > calendario_completo['CAPACIDAD_ESTIMADA']

# Mostrar los registros donde se excede la capacidad
excede_capacidad = calendario_completo[calendario_completo['EXCEDE_CAPACIDAD'] == True]
print("\nRegistros donde el número de estudiantes excede la capacidad estimada del aula:")
display(excede_capacidad)

# Calcular el porcentaje de ocupación
calendario_completo['PORCENTAJE_OCUPACION'] = (calendario_completo['IND_ALUMNOS_GRUPO_REAL'] / calendario_completo['CAPACIDAD_ESTIMADA'] * 100)

# Verificar solapamientos
solapamientos = calendario_completo.groupby(['ID_UBICACION', 'ID_FECHA_GRUPO', 'ID_QUART_HORA']).size().reset_index(name='CONTEO')
solapamientos = solapamientos[solapamientos['CONTEO'] > 1]
print("\nPosibles solapamientos de horarios en una misma aula:")
display(solapamientos)
```

### 7. Visualización de Ocupación por Franja Horaria

- Calcula la ocupación media de estudiantes por franja horaria.
- Genera un gráfico de barras para visualizar los horarios con mayor y menor ocupación.
- Utiliza esta información para identificar patrones temporales

```python
# Calcular la ocupación media por franja horaria
ocupacion_franja = calendario_completo.groupby('ID_QUART_HORA')['IND_ALUMNOS_GRUPO_REAL'].mean().reset_index()

# Ordenar por ID_QUART_HORA para una mejor visualización
ocupacion_franja = ocupacion_franja.sort_values('ID_QUART_HORA')

# Visualizar la ocupación media por franja horaria
plt.figure(figsize=(12,6))
plt.bar(ocupacion_franja['ID_QUART_HORA'], ocupacion_franja['IND_ALUMNOS_GRUPO_REAL'], color='skyblue')
plt.xlabel('ID Cuarto de Hora')
plt.ylabel('Ocupación Media de Estudiantes')
plt.title('Ocupación Media por Franja Horaria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Opcional: Mostrar los valores numéricos
print("\nOcupación media por franja horaria:")
display(ocupacion_franja)
```
