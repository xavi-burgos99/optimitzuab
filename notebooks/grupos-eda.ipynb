## Data Cleaning

We are only interested in the data for the academic year 2024. Let's filter the data and take a look at the first few rows.

```python
 # Take only ID_CURSO_ACADEMICO=2024
df_2024 = df_sorted[df_sorted['ID_CURSO_ACADEMICO'] == 2024]
df_2024.head()
```

The ID_SEMESTRE column is no longer needed. Let's drop it and take a look at the first few rows. With ID_PERIODO_DOCENTE we can count the number of semesters.

```python
# Drop ID_SEMESTRE as it is no longer used
df_2024 = df_2024.drop(columns=['ID_SEMESTRE'])
```

```python
df_2024.head()
```

```python
# Print number of entries
print(f'Number of entries: {len(df_2024)}')

# We can use the ID_PERIODO_DOCENTE to count the number of semesters
print(f'Number of semesters: {len(df_2024.ID_PERIODO_DOCENTE.unique())}')
```

We encode the ID_PERIODO_DOCENTE into numeric values and take a look at the first few rows. This is done to simplify the data and remove categorical values.

```python
# Encode periodo docente into 0, 1, 2, numeric
df_2024['ID_PERIODO_DOCENTE'] = pd.Categorical(df_2024['ID_PERIODO_DOCENTE'])
df_2024['ID_PERIODO_DOCENTE'] = df_2024['ID_PERIODO_DOCENTE'].cat.codes
df_2024.head()
```

```python
# number of gropus per each periodo docente
df_2024.groupby('ID_PERIODO_DOCENTE').size()
```

```python
# check if ID_GROUP -1 exists
if -1 in df_2024['ID_GRUPO'].values:
    print(f'Group ID {-1} exists in the dataset.')
else:
    print(f'Group ID {-1} does not exist in the dataset.')
```

We don't need ID_IDIOMA, ID_UNIDAD_DOCENTE, ID_DEPARTAMENT_DELEG_DECODA, ID_VALOR_PERIODO_DOC. Let's drop them and take a look at the last few rows.

```python
# Drop ID_IDIOMA
df_2024 = df_2024.drop(columns=
                       ["IDIOMA_MAJORITARI", 
                        "ID_IDIOMA", 
                        "ID_UNIDAD_DOCENTE", 
                        "ID_DEPARTAMENT_DELEG_DECODA",
                        "ID_VALOR_PERIODO_DOC"])
```

```python
df_2024.reset_index()
df_2024.tail()
```

```python
# Print number of entries
print(f'Number of entries: {len(df_2024)}')

# We can use the ID_PERIODO_DOCENTE to count the number of semesters
print(f'Number of codigo grupo: {len(df_2024.ID_COD_GRUPO.unique())}')
```

```python
# Print all tipos docencia unique values and their counts
print(df_2024['ID_TIPO_DOCENCIA'].value_counts())
print(len(df_2024['ID_TIPO_DOCENCIA'].unique()))
```

Drop ID_CENTRO_DEPART as it is no longer needed and take a look at the first few rows.

```python
df_2024 = df_2024.drop(columns=['ID_CENTRO_DEPART'])
```

```python
df_2024.head()
```

Keep only rows with IND_ALUMNOS_GRUPO_REAL > 0 and IND_HORAS_PREVISTAS > 0

```python
# Keep only rows with IND_ALUMNOS_GRUPO_REAL > 0 and IND_HORAS_PREVISTAS > 0
df_2024 = df_2024[(df_2024['IND_ALUMNOS_GRUPO_REAL'] > 0)]
df_2024 = df_2024[(df_2024['IND_HORAS_PREVISTAS'] > 0)]
```
